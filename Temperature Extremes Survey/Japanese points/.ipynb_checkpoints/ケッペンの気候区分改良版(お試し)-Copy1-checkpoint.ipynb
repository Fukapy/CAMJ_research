```python
#### ケッペンの気候区分
#2021.10.19
# https://qiita.com/motoki1990/items/0274d8bcf1a97fe4a869
#csv_input = pd.read_csv(filepath_or_buffer="D:\株価情報CSV\TEST_STOCK.csv", encoding="ms932", sep=",")

#データ読み込み


#地域の入力
place=input("地域を入力してください=")
#気温と降水量の入力
Jant=float(input("1月の平均気温(℃)="))
Febt=float(input("2月の平均気温(℃)="))
Mart=float(input("3月の平均気温(℃)="))
Aprt=float(input("4月の平均気温(℃)="))
Mayt=float(input("5月の平均気温(℃)="))
Junt=float(input("6月の平均気温(℃)="))
Jult=float(input("7月の平均気温(℃)="))
Augt=float(input("8月の平均気温(℃)="))
Sept=float(input("9月の平均気温(℃)="))
Octt=float(input("10月の平均気温(℃)="))
Novt=float(input("11月の平均気温(℃)="))
Dect=float(input("12月の平均気温(℃)="))
Janr=float(input("1月の降水量(mm)="))
Febr=float(input("2月の降水量(mm)="))
Marr=float(input("3月の降水量(mm)="))
Aprr=float(input("4月の降水量(mm)="))
Mayr=float(input("5月の降水量(mm)="))
Junr=float(input("6月の降水量(mm)="))
Julr=float(input("7月の降水量(mm)="))
Augr=float(input("8月の降水量(mm)="))
Sepr=float(input("9月の降水量(mm)="))
Octr=float(input("10月の降水量(mm)="))
Novr=float(input("11月の降水量(mm)="))
Decr=float(input("12月の降水量(mm)="))

#年平均気温T(℃)
T=int((Jant+Febt+Mart+Aprt+Mayt+Junt+Jult+Augt+Sept+Octt+Novt+Dect)*10/12)/10
#夏雨型と冬雨型の決定
if Janr<=Julr:
    rstyle=0#summer
else:
    rstyle=1#winter
#最多雨月降水量Rw(mm)と最小雨月降水量Rd(mm)
Rw=max(Janr,Febr,Marr,Aprr,Mayr,Junr,Julr,Augr,Sepr,Octr,Novr,Decr)
Rd=min(Janr,Febr,Marr,Aprr,Mayr,Junr,Julr,Augr,Sepr,Octr,Novr,Decr)
#乾燥限界指数r(mm)
if rstyle==0 and Rw>=10*Rd:
    r=20*(T+14)
elif rstyle==1 and Rw>=3*Rd:
    r=20*T
else:
    r=20*(T+7)#多雨型
#年降水量R(mm)
R=int((Janr+Febr+Marr+Aprr+Mayr+Junr+Julr+Augr+Sepr+Octr+Novr+Decr)*10)/10
#最暖月平均気温Tw(℃)と最寒月平均気温Tc(℃)
Tw=max(Jant,Febt,Mart,Aprt,Mayt,Junt,Jult,Augt,Sept,Octt,Novt,Dect)
Tc=min(Jant,Febt,Mart,Aprt,Mayt,Junt,Jult,Augt,Sept,Octt,Novt,Dect)

#気候区分を判別するプログラム
if 1/2*r>=R:
    climate=5#砂漠気候
    print("BW")
elif r>R>=1/2*r:
    climate=4#ステップ気候
    print("BS")
elif R>r:
    climate=100#湿潤気候帯
    if Tc>=18:
        climate=101#熱帯
        if Rd>=60:
            climate=1
            print("Af")
        elif Rd<60 and Rd>=-0.04*R+100:
            climate=2
            print("Am")
        else:
            climate=3
            print("Aw")
            
    elif 18>Tc>=-3:
        climate=102#温帯
        if rstyle==0 and Rw>=10*Rd:
            climate=7
            print("Cw")#温暖冬季少雨気候
        elif rstyle==1 and Rw>=3*Rd:
            climate=6
            print("Cs")#地中海性気候
        else:
            climate=104#Cf
            if Tw>=22:
                climate=8
                print("Cfa")#温暖湿潤気候
            else:
                climate=9
                print("Cfb")#西岸海洋性気候    
            
    elif -3>Tc and Tw>=10:
        climate=103#冷帯
        if rstyle==0 and Rw>=10*Rd:
            climate=11
            print("Dw")#冷帯冬季少雨気候
        else:
            climate=10
            print("Df")#冷帯湿潤気候   
            
    elif 10>Tw>=0:
        climate=12
        print("ET")#ツンドラ気候
    elif 0>Tw:
        climate=13
        print("EF")#氷雪気候

#
```

```python
#データの読み込み
import numpy as np
import pandas as pd
csv_input = pd.read_csv("climat_47407_201901-202012.csv", encoding="ms932")

Jant=float(csv_input.values[2, 3])
Febt=float(csv_input.values[3, 3])
Mart=float(csv_input.values[4, 3])
Aprt=float(csv_input.values[5, 3])
Mayt=float(csv_input.values[6, 3])
Junt=float(csv_input.values[7, 3])
Jult=float(csv_input.values[8, 3])
Augt=float(csv_input.values[9, 3])
Sept=float(csv_input.values[10, 3])
Octt=float(csv_input.values[11, 3])
Novt=float(csv_input.values[12, 3])
Dect=float(csv_input.values[13, 3])
Janr=float(csv_input.values[2, 4])
Febr=float(csv_input.values[3, 4])
Marr=float(csv_input.values[4, 4])
Aprr=float(csv_input.values[5, 4])
Mayr=float(csv_input.values[6, 4])
Junr=float(csv_input.values[7, 4])
Julr=float(csv_input.values[8, 4])
Augr=float(csv_input.values[9, 4])
Sepr=float(csv_input.values[10, 4])
Octr=float(csv_input.values[11, 4])
Novr=float(csv_input.values[12, 4])
Decr=float(csv_input.values[13, 4])

mat=np.matrix([[Jant,Febt,Mart,Aprt,Mayt,Junt,Jult,Augt,Sept,Octt,Novt,Dect],[Janr,Febr,Marr,Aprr,Mayr,Junr,Julr,Augr,Sepr,Octr,Novr,Decr]])
print(mat)



def keppen(Jant,Febt,Mart,Aprt,Mayt,Junt,Jult,Augt,Sept,Octt,Novt,Dect,Janr,Febr,Marr,Aprr,Mayr,Junr,Julr,Augr,Sepr,Octr,Novr,Decr):
    #年平均気温T(℃)
    T=int((Jant+Febt+Mart+Aprt+Mayt+Junt+Jult+Augt+Sept+Octt+Novt+Dect)*10/12)/10
    #夏雨型と冬雨型の決定
    if Janr<=Julr:
        rstyle=0#summer
    else:
        rstyle=1#winter
    #最多雨月降水量Rw(mm)と最小雨月降水量Rd(mm)
    Rw=max(Janr,Febr,Marr,Aprr,Mayr,Junr,Julr,Augr,Sepr,Octr,Novr,Decr)
    Rd=min(Janr,Febr,Marr,Aprr,Mayr,Junr,Julr,Augr,Sepr,Octr,Novr,Decr)
    #乾燥限界指数r(mm)
    if rstyle==0 and Rw>=10*Rd:
        r=20*(T+14)
    elif rstyle==1 and Rw>=3*Rd:
        r=20*T
    else:
        r=20*(T+7)#多雨型
    #年降水量R(mm)
    R=int((Janr+Febr+Marr+Aprr+Mayr+Junr+Julr+Augr+Sepr+Octr+Novr+Decr)*10)/10
    #最暖月平均気温Tw(℃)と最寒月平均気温Tc(℃)
    Tw=max(Jant,Febt,Mart,Aprt,Mayt,Junt,Jult,Augt,Sept,Octt,Novt,Dect)
    Tc=min(Jant,Febt,Mart,Aprt,Mayt,Junt,Jult,Augt,Sept,Octt,Novt,Dect)

    #気候区分を判別するプログラム
    if 1/2*r>=R:
        climate=5#砂漠気候
        print("BW:砂漠気候")
    elif r>R>=1/2*r:
        climate=4#ステップ気候
        print("BS:ステップ気候")
    elif R>r:
        climate=100#湿潤気候帯
        if Tc>=18:
            climate=101#熱帯
            if Rd>=60:
                climate=1
                print("Af:熱帯雨林気候")
            elif Rd<60 and Rd>=-0.04*R+100:
                climate=2
                print("Am:熱帯モンスーン気候")
            else:
                climate=3
                print("Aw:サバナ気候")
            
        elif 18>Tc>=-3:
            climate=102#温帯
            if rstyle==0 and Rw>=10*Rd:
                climate=7
                print("Cw:温暖冬季少雨気候")#温暖冬季少雨気候
            elif rstyle==1 and Rw>=3*Rd:
                climate=6
                print("Cs:地中海性気候")#地中海性気候
            else:
                climate=104#Cf
                if Tw>=22:
                    climate=8
                    print("Cfa:温暖湿潤気候")#温暖湿潤気候
                else:
                    climate=9
                    print("Cfb:西岸海洋性気候")#西岸海洋性気候    
            
        elif -3>Tc and Tw>=10:
            climate=103#冷帯
            if rstyle==0 and Rw>=10*Rd:
                climate=11
                print("Dw:冷帯冬季少雨気候")#冷帯冬季少雨気候
            else:
                climate=10
                print("Df:冷帯湿潤気候")#冷帯湿潤気候   
            
        elif 10>Tw>=0:
            climate=12
            print("ET:ツンドラ気候")#ツンドラ気候
        elif 0>Tw:
            climate=13
            print("EF:氷雪気候")#氷雪気候  
    
    return

keppen(Jant,Febt,Mart,Aprt,Mayt,Junt,Jult,Augt,Sept,Octt,Novt,Dect,Janr,Febr,Marr,Aprr,Mayr,Junr,Julr,Augr,Sepr,Octr,Novr,Decr)
```
