```python
#2021.10.24

#データの読み込み
import numpy as np
import pandas as pd
import math
csv_input = pd.read_csv("climat_47662_201901-202012.csv", encoding="ms932")

def saisyobi():

    Jant=float(csv_input.values[2, 3])
    Febt=float(csv_input.values[3, 3])
    Mart=float(csv_input.values[4, 3])
    Aprt=float(csv_input.values[5, 3])
    Mayt=float(csv_input.values[6, 3])
    Junt=float(csv_input.values[7, 3])
    Jult=float(csv_input.values[8, 3])
    Augt=float(csv_input.values[9, 3])
    Sept=float(csv_input.values[10, 3])
    Octt=float(csv_input.values[11, 3])
    Novt=float(csv_input.values[12, 3])
    Dect=float(csv_input.values[13, 3])

#最暖月の判定
    if Aprt>=Mayt:
        Twm=4
    elif Mayt>=Junt:
        Twm=5
    elif Junt>=Jult:
        Twm=6
    elif Jult>=Augt:
        Twm=7
    elif Augt>=Sept:
        Twm=8
    elif Sept>=Octt:
        Twm=9
    elif Octt>=Novt:
        Twm=10
    else:
        Twm=11

#最暖月の周辺のデータを抽出
    T_bM=float(csv_input.values[Twm, 3])
    T_M=float(csv_input.values[Twm+1, 3])
    T_aM=float(csv_input.values[Twm+2, 3])

#極値となる日
    if T_bM==T_M:
        Twd=1
    elif T_bM==T_aM:
        Twd=16
    elif T_bM<T_aM:
        Twd=(0.5+(T_aM-T_bM)/(T_M-T_bM)/2)*30
    elif T_bM>T_aM:
        Twd=(0.5-(T_bM-T_aM)/(T_M-T_aM)/2)*30
    Twd_int=int(Twd)

#最も暑い日の出力
    print("最暑日(直線近似)は",Twm,"月",Twd_int,"日です。")

#極大　放物線(ここから2021.12.12)
    x_M=(T_M-T_aM+(Twm+1/2)*(T_aM-2*T_M+T_bM))/(T_aM-2*T_M+T_bM)
    f,i= math.modf(x_M+0.5)
    print("最暑日(放物線近似)は",int(i),"月",int(30*f),"日です。")
    
    return


#2021.12.12
def saikanbi():

    Jant=float(csv_input.values[2, 3])
    Febt=float(csv_input.values[3, 3])
    Mart=float(csv_input.values[4, 3])
    Aprt=float(csv_input.values[5, 3])
    Mayt=float(csv_input.values[6, 3])
    Junt=float(csv_input.values[7, 3])
    Jult=float(csv_input.values[8, 3])
    Augt=float(csv_input.values[9, 3])
    Sept=float(csv_input.values[10, 3])
    Octt=float(csv_input.values[11, 3])
    Novt=float(csv_input.values[12, 3])
    Dect=float(csv_input.values[13, 3])

#最寒月の判定
    if Dect<=Jant:
        Tcm=12
    elif Jant<=Febt:
        Tcm=1
    elif Febt<=Mart:
        Tcm=2
    elif Mart<=Aprt:
        Tcm=3
    else:
        Tcm=4
    
#最寒月の周辺のデータを抽出
    if Tcm==1:
        T_bm=float(csv_input.values[13, 3])
        T_m=float(csv_input.values[2, 3])
        T_am=float(csv_input.values[3, 3])
    else:
        T_bm=float(csv_input.values[Tcm, 3])
        T_m=float(csv_input.values[Tcm+1, 3])
        T_am=float(csv_input.values[Tcm+2, 3])

        
    print(T_bm,T_m,T_am)

#極値となる日
    if T_bm==T_m:
        Tcd=1
    elif T_bm==T_am:
        Tcd=15
    elif T_bm<T_am:
        Tcd=(0.5+(T_am-T_bm)/(T_bm-T_m+0.0001)/2)*30
    elif T_bm>T_am:
        Tcd=(0.5-(T_bm-T_am)/(T_am-T_m+0.0001)/2)*30
    print(Tcd)
    Tcd_int=int(Tcd)
#最も寒い日の出力
    print("最寒日(直線近似)は",Tcm,"月",Tcd_int,"日です。")

#極小　放物線(ここから2021.12.12)
    x_m=(T_m-T_am+(Tcm+1/2)*(T_am-2*T_m+T_bm))/(T_am-2*T_m+T_bm)
    f,i= math.modf(x_m+0.5)
    print("最寒日(放物線近似)は",int(i),"月",int(30*f),"日です。")
    
    return
```
