```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("気象2021ヒストグラム用.csv")
```

```python
data
```

```python
# df = pd.read_csv("mmchars.csv", names=['Vol', 'chars'])
# print(df.describe())
plt.hist(data['極小誤差(直線)'], bins=10)
plt.xlim(0, 20)
plt.title('min(liner)')
plt.xlabel('days')
plt.ylabel('count')
plt.show()
```

```python
plt.hist(data['極大誤差(直線)'], bins=10)
plt.xlim(0, 20)
plt.title('max(liner)')
plt.xlabel('days')
plt.ylabel('count')
plt.show()
```

```python
plt.hist(data['極小誤差(放物線)'], bins=10)
plt.xlim(0, 20)
plt.title('min(parabolic)')
plt.xlabel('days')
plt.ylabel('count')
plt.show()
```

```python
plt.hist(data['極大誤差(放物線)'], bins=10)
plt.xlim(0, 20)
plt.title('max(parabolic)')
plt.xlabel('days')
plt.ylabel('count')
plt.show()
```
